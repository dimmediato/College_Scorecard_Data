```python
import numpy as np
import pandas as pd
import seaborn as sns
import scipy.stats as ss
import matplotlib.pyplot as plt

Texas = pd.read_csv("~\\Documents\\NYCDSA\\First Project\\Git_Workspace\\College_Scorecard_Data\\Data\\Texas_Retention.csv")
Texas['Change in GPA'] = Texas['Change in GPA'].str.replace('No Chang', 'No Change')
Texas
```

```python
Overall = sns.boxplot(x = 'Retention', y = 'Overall GPA', palette = ("#c37ef0","#0096FF"), hue = 'Gender', data = Texas)
Overall.set(title = 'Does GPA Impact Retention?')
```

```python
HS = sns.boxplot(x = 'Retention', y = 'High School GPA', palette = ("#c37ef0","#0096FF"), hue = 'Gender', data = Texas)
HS = HS.legend(loc='upper right', title = 'Gender', facecolor='white', framealpha=1)
```

```python
HS_violin = sns.violinplot(x = 'Retention', y = 'High School GPA', palette = ("#c37ef0","#0096FF"), hue = 'Gender', data = Texas)
HS_violin= HS_violin.legend(loc='upper right', title = 'Gender', facecolor='white', framealpha=1)
```

```python
RCP = {
    'Negative': '#ff0000',
    'Positive': '#228B22',
    'No Change': 'tab:blue',
}

Retention_Change = sns.histplot(binwidth=1,
            x="Retention",
            hue="Change in GPA" ,
            data=Texas, palette = RCP,
            stat="count",
            multiple="dodge").set(title= "Does Change in GPA Impact Retention?", ylabel='Number of Students')
```

```python
sns.histplot(binwidth=1,
            x="Retention",
            hue="Received Financial Aid", 
            palette = ("#228B22", "#ff0000"),
            data=Texas,
            stat="count",
            multiple="dodge").set(title= "Does Receiving Financial Aid Impact Student Retention?", ylabel='Number of Students')
```
